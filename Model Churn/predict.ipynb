```python
import mlflow.client
import pandas as pd
import mlflow
```

```python
mlflow.set_tracking_uri("http://localhost:5000/")
```

```python
client = mlflow.client.MlflowClient()
version = max([int(x.version) for x in client.search_model_versions("name='churn_model'")])
```

```python
model = mlflow.sklearn.load_model(f"models:/churn_model/{version}")
```

```python
df = pd.read_csv("data/abt.csv", sep=",")
df
```

```python
X = df.head(10)[model.feature_names_in_]
proba = model.predict_proba(X)
proba
```
